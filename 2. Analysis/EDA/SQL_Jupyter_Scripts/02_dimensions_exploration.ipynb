<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<b>1. Retrieve a list of unique countries from which customers originate</b>
</div>


```python
query = """
SELECT DISTINCT 
    country 
FROM gold.dim_customers
ORDER BY country;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```

<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<strong>2. Retrieve a list of unique categories, subcategories, and products</strong>
</div>


```python
query = """
SELECT 
    DISTINCT TOP 10
    category, 
    subcategory, 
    product_name 
FROM gold.dim_products
ORDER BY category, subcategory, product_name;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```
