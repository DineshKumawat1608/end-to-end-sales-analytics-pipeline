<h2 style="margin-top:0; color:#000;">Magnitude Analysis</h2>

<h4 style="color:#000;">Purpose</h4>
<ul>
  <li>Quantify data and group results by specific dimensions.</li>
  <li>Understand data distribution across different categories.</li>
</ul>

<h4 style="color:#000;">SQL Functions Used</h4>
<ul>
  <li><b>SUM()</b>, <b>COUNT()</b>, <b>AVG()</b></li>
  <li><b>GROUP BY</b>, <b>ORDER BY</b></li>
</ul>

</div>


<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<b>1. Find total customers by countries</b>
</div>

```python
query = """
SELECT
    country,
    COUNT(customer_key) AS total_customers
FROM gold.dim_customers
GROUP BY country
ORDER BY total_customers DESC;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```

<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<strong>2. Find total customers by gender</strong>
</div>

```python
query = """
SELECT
    gender,
    COUNT(customer_key) AS total_customers
FROM gold.dim_customers
GROUP BY gender
ORDER BY total_customers DESC;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```

<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<strong>3. Find total products by category</strong>
</div>

```python
query = """
SELECT
    category,
    COUNT(product_key) AS total_products
FROM gold.dim_products
GROUP BY category
ORDER BY total_products DESC;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```

<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<strong>4. What is the average costs in each category?</strong>
</div>

```python
query = """
SELECT
    category,
    AVG(cost) AS avg_cost
FROM gold.dim_products
GROUP BY category
ORDER BY avg_cost DESC;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```

<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<strong>5. What is the total revenue generated for each category?</strong>
</div>

```python
query = """
SELECT
    p.category,
    SUM(f.sales_amount) AS total_revenue
FROM gold.fact_sales f
LEFT JOIN gold.dim_products p
    ON p.product_key = f.product_key
GROUP BY p.category
ORDER BY total_revenue DESC;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```

<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<strong>6. What is the total revenue generated by each customer?</strong>
</div>

```python
query = """
SELECT TOP 10
    c.customer_key,
    c.first_name,
    c.last_name,
    SUM(f.sales_amount) AS total_revenue
FROM gold.fact_sales f
LEFT JOIN gold.dim_customers c
    ON c.customer_key = f.customer_key
GROUP BY 
    c.customer_key,
    c.first_name,
    c.last_name
ORDER BY total_revenue DESC;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```

<div style="background-color:#f4f8ff; padding:10px; border-left:6px solid #1f4fd8; border-radius:6px; color:#000; font-size:20px;">
<strong>7. What is the distribution of sold items across countries?</strong>
</div>

```python
query = """
SELECT
    c.country,
    SUM(f.quantity) AS total_sold_items
FROM gold.fact_sales f
LEFT JOIN gold.dim_customers c
    ON c.customer_key = f.customer_key
GROUP BY c.country
ORDER BY total_sold_items DESC;
"""

df = pd.read_sql(query, engine)
display(HTML(df.to_html(index=False)))
```
